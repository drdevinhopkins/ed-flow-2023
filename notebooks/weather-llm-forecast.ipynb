```python
import pandas as pd
import numpy as np

def forecast_to_future_metar(forecast_df):
    """
    Convert Environment Canada hourly forecast → CYUL METAR-like rows
    matching the exact schema of your metar_df.
    
    forecast_df must contain:
    ['period','condition','temperature','precip_probability','wind_speed','wind_direction']
    """

    df = forecast_df.copy()

    # --- REQUIRED: METAR SCHEMA OUTPUT ---
    cols = [
        'station','valid','tmpf','dwpf','relh','drct','sknt','p01i','alti','mslp',
        'vsby','gust','skyc1','skyc2','skyc3','skyc4','skyl1','skyl2','skyl3','skyl4',
        'wxcodes','ice_accretion_1hr','ice_accretion_3hr','ice_accretion_6hr',
        'peak_wind_gust','peak_wind_drct','peak_wind_time','feel','metar','snowdepth'
    ]

    # --- TIMESTAMP ---
    df["valid"] = pd.to_datetime(df["period"]).dt.tz_convert("America/Montreal")

    # --- STATION ---
    df["station"] = "CYUL"

    # --- TEMPERATURE (°F because tmpf is Fahrenheit in IEM METAR feeds) ---
    df["tmpf"] = df["temperature"] * 9/5 + 32

    # --- DEWPOINT (approx) ---
    df["dwpf"] = df["tmpf"] - 4  # simple heuristic or replace with LLM

    # --- RELATIVE HUMIDITY (estimate using temp/dp) ---
    # Using simplified RH formula
    rh = 100 * np.exp((17.625*(df["dwpf"]-32)*5/9)/(243.04+((df["dwpf"]-32)*5/9))) / \
               np.exp((17.625*(df["tmpf"]-32)*5/9)/(243.04+((df["tmpf"]-32)*5/9)))
    df["relh"] = np.clip(rh, 20, 100).astype(int)

    # --- WIND ---
    df["drct"] = df["wind_direction"]              # degrees
    df["sknt"] = df["wind_speed"] / 1.852 * 1.94384   # km/h → knots
    df["sknt"] = df["sknt"].round(0)

    # --- PRECIPITATION INTENSITY ESTIMATE (p01i = precip last hour, inches) ---
    df["p01i"] = np.where(df["precip_probability"] >= 60, 0.01, 0.00)

    # --- ALTIMETER (inHg) and MSLP (hPa) ---
    df["mslp"] = 1012.0  # climatological mean
    df["alti"] = df["mslp"] * 0.02953  # hPa → inHg

    # --- VISIBILITY (statute miles) ---
    def estimate_vis(row):
        cond = row["condition"].lower()
        if "snow" in cond:
            return 1.0
        if "rain" in cond:
            return 2.0
        if "fog" in cond or "mist" in cond:
            return 0.5
        return 6.0
    df["vsby"] = df.apply(estimate_vis, axis=1)

    # --- GUST ---
    df["gust"] = 0.0

    # --- CLOUD LAYERS (sky cover amount + height) ---
    # For forecasting, keep simple deterministic placeholders:
    df["skyc1"] = "BKN"
    df["skyc2"] = df["skyc3"] = df["skyc4"] = ""
    df["skyl1"] = 2500   # feet AGL
    df["skyl2"] = df["skyl3"] = df["skyl4"] = np.nan

    # --- WEATHER CODES (METAR format) ---
    def condition_to_code(cond):
        c = cond.lower()
        if "snow" in c:
            return "-SN"
        if "rain" in c:
            return "-RA"
        if "mix" in c or ("snow" in c and "rain" in c):
            return "-RASN"
        if "fog" in c or "mist" in c:
            return "BR"
        return ""
    df["wxcodes"] = df["condition"].apply(condition_to_code)

    # --- ICE ACCRETION ---
    df["ice_accretion_1hr"] = 0.0
    df["ice_accretion_3hr"] = 0.0
    df["ice_accretion_6hr"] = 0.0

    # --- PEAK WIND (optional, set missing) ---
    df["peak_wind_gust"] = np.nan
    df["peak_wind_drct"] = np.nan
    df["peak_wind_time"] = pd.NaT

    # --- FEELS-like TEMP (placeholder) ---
    df["feel"] = df["tmpf"]

    # --- RAW METAR STRING (optional; leave empty or LLM-generated later) ---
    df["metar"] = ""

    # --- SNOW DEPTH ---
    df["snowdepth"] = np.nan

    # --- RETURN EXACT SCHEMA ---
    return df[cols]
```
