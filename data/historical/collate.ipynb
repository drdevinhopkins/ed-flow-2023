```python
import pandas as pd
import datetime
```

```python
output = pd.DataFrame()
df = pd.read_csv('2021q1.csv')
df["ds"] = pd.to_datetime(
    df["dateflg"] + " " + (df["timeflg"] - 1).astype(str) + ":00", format='mixed') + datetime.timedelta(hours=1)
df = df.set_index('ds').reset_index().drop(['dateflg','timeflg'], axis=1)
output = pd.concat([output, df], ignore_index=True)
output
```

```python
output = pd.DataFrame()
for file in ['2021q1.csv', '2021q2.csv', '2021q3.csv', '2021q4.csv', '2022q1.csv', '2022q2.csv', '2022q3.csv', '2022q4.csv']:
    df = pd.read_csv(file)
    df["ds"] = pd.to_datetime(
        df["dateflg"] + " " + (df["timeflg"] - 1).astype(str) + ":00", format='mixed') + datetime.timedelta(hours=1)
    df = df.set_index('ds').reset_index().drop(['dateflg','timeflg'], axis=1)
    output = pd.concat([output, df], ignore_index=True)
df = pd.read_csv('2023data.csv')
df.ds = pd.to_datetime(df.ds)
output = pd.concat([output, df], ignore_index=True)
output = output.drop_duplicates(keep='first').sort_values(by='ds', ascending=True)
output
```

```python
output.to_csv('allData.csv', index=False)
```
