```python
import pandas as pd
hist_weather = pd.read_csv("montreal_weather_2021_2025.csv")
hist_weather.head()
```

```python
#rename Date/Time (LST) to ds and convert to pandas datetime
hist_weather = hist_weather.rename(columns={"Date/Time (LST)": "ds"})
hist_weather['ds'] = pd.to_datetime(hist_weather['ds'], errors='coerce')
hist_weather = hist_weather.sort_values('ds')
hist_weather.head()
```

```python
hist_weather.to_csv("montreal_weather_2021_2025_cleaned.csv", index=False)
```
